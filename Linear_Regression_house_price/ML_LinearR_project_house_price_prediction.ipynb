INTRODUCTION

House price prediction is a crucial application of machine learning that leverages data-driven techniques to estimate property values based on factors such as location, size, and market trends.

IMPORTING LIBRARIES AND LOADING DATA

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
data=pd.read_csv(r"C:\Users\Lenovo\Documents\Data analytics\data_analytics\project\house_price_dataset.csv")
data.head(10)
```

EXTRACTING INDEPENDENT AND DEPENDENT VARIABLE

```python
x=data.iloc[:,:7]
x=pd.DataFrame(x)
y=data['House_Price']
y=pd.DataFrame(y)
```

SPLITING DATASET INTO TRAIN AND TEST DATA AND MODEL BUILDING

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.2,random_state=67)
model=LinearRegression()
model.fit(x_train,y_train)
```

PREDICTING THE TEST SET RESULT

```python
y_predict=model.predict(x_test)
print(y_test)
print(y_predict)
```

EVALUATE PREDICTIONS ACCURACY

```python
mse_value=metrics.mean_squared_error(y_test,y_predict)
print(mse_value)
accuracy=metrics.r2_score(y_test,y_predict)
print(accuracy*100)
```

SUMMARY

The linear regression emerged as the most accurate model for house price prediction.
